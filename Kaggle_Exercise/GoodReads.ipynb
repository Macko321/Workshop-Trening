```python
sns.set_context(context='talk')
plt.figure(figsize=(20,15))
books = df_1['title'].value_counts()[:20]
number_of_pages = df_1.text_reviews_count[:20]
sns.barplot(x = books, y = books.index, palette = 'deep')
plt.show()
```

```python
sns.set_context(context='talk')
plt.figure(figsize=(20,15))
books = df_1['title'].value_counts()[:20]
review = df_1.text_reviews_count[:20]
sns.barplot(x = review, y = review.index, palette = 'deep')
plt.show()
```

```python
sns.set_context(context='poster')
plt.figure(figsize=(40,50))
authors = df_1['authors'].value_counts()[:30]

sns.barplot(x = authors, y = authors.index, palette = 'deep')

plt.title("Authors with most books")
plt.xlabel("Number of books")
plt.ylabel("Authors")
plt.show()
```
